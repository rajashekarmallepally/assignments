### The min and max temperatures in a city In India for each months of the year.We would like to find a function to describe this and show it graphically,

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#temperature data

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

# ploting using matplotlib
months = np.arange(12)
plt.plot(months, temp_max, 'm^')
plt.plot(months, temp_min, 'gv')
plt.xlabel('Month')
plt.ylabel('Min and Max temperature')
```

#### Fitting it to a periodic function




```python
from scipy import optimize

#funtion to describe min and max temperatures
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      temp_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      temp_min, [-40, 20, 0])
```

#### Plotting the fit




```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, temp_max, 'm^')
plt.plot(days, yearly_temps(days, *res_max), 'm-')
plt.plot(months, temp_min, 'gv')
plt.plot(days, yearly_temps(days, *res_min), 'g-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
